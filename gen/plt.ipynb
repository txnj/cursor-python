### matplotlib简单示例

```python
import matplotlib.pyplot as plt

# 数据
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.rcParams['font.family'] = 'Microsoft JhengHei'  # 或者其他支持中文的字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
# 创建图形
plt.figure()

# 绘制折线图
plt.plot(x, y, label='折线图', marker='o')

# 添加标题和标签
plt.title('简单折线图示例')
plt.xlabel('X轴')
plt.ylabel('Y轴')

# 显示图例
plt.legend()

# 显示图形
plt.show()
```
